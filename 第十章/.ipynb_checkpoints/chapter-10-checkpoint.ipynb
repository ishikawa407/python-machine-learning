### 10.3.1 梯度下降法求解回归参数
            ols成本函数与Adaline的成本函数相同 平方和误差（SSE）

```python
# 以住房数据RM为作为解释变量 预测MEDV的模型
```

```python
df = pd.read_csv('housing.data.txt', header=None, sep='\s+')
df.columns = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX',
              'RM', 'AGE', 'DIS', 'RAD', 'TAX',
              'PTRATIO', 'B', 'LSTAT', 'MEDV']
df.iloc[:5,:]
```

```python
X_1, y[:5]
```

```python
class LinearRegressionGD(object):

    def __init__(self, eta=0.001, n_iter=20):
        self.eta = eta
        self.n_iter = n_iter

    def fit(self, X, y):
        
        rgen = np.random.RandomState(1)
        self.w_ = np.zeros(1 + X.shape[1])
        self.w_ = rgen.normal(loc=0.0, scale=0.01,
                              size=1 + X.shape[1])
        self.cost_ = []
        self.errors_ = []

        for i in range(self.n_iter):
            output = self.net_input(X)
            errors = (y - output)
            self.w_[1:] += self.eta * X.T.dot(errors)
            self.w_[0] += self.eta * errors.sum()
            cost = (errors**2).sum() / 2.0
            self.cost_.append(cost)
            self.errors_.append(errors.sum()) 
        return self

    def net_input(self, X):
        return np.dot(X, self.w_[1:]) + self.w_[0]
    
    def predict(self, X):
        return self.net_input(X)
```

```python
# 先对数据进行标准化
X = df[['RM']].values
y = df['MEDV'].values
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
sc_y = StandardScaler()
X_std = sc_x.fit_transform(X)
y_std = sc_y.fit_transform(y[:, np.newaxis]).flatten()
```

```python
plt.scatter(X_std, y_std)
plt.show()
```

```python
lr = LinearRegressionGD()
```

```python
lr.fit(X_std, y_std)
lr.w_
lr.errors_
```

```python
sns.reset_orig() # reset matplotlib style
```

```python
plt.plot(range(1, lr.n_iter+1), lr.cost_)
plt.ylabel('SSE')
plt.xlabel('Epoch')
plt.show()
```

```python
# 观察线性回归与训练数据的吻合程度
# 定义简单辅助函数Lin_regplot绘制训练样本的散点图并添加回归线
def lin_regplot(X, y, model):
    plt.scatter(X, y, c='steelblue', edgecolors='white', s=70)
    plt.plot(X, model.predict(X), color='black', lw=2)
```

```python
# 利用lin_regplot绘制房间书和房价的关系图
lin_regplot(X_std, y_std, lr)
plt.xlabel('平均房间数目[RM]（标准化）')
plt.ylabel('房价/1000美元（标准化）')
plt.show()
```

## 10.4 利用RANSAC拟合稳健的回归模型

          总结迭代RANSAC算法：
          1. 随机选择一定数量的样本作为内点来拟合模型
          2. 用模型测试所有其他的点，把落在用户给定容限范围内的点放入内点集
          3. 调整模型中使用的所有的内点
          4. 用内点重新拟合模型
          5. 评估模型预测结果与内点集相比较的误差
          6. 如果性能达到用户定义的阈值或制定的叠代数，则终止；否则返回1

```python
# 用scikit——learn的RANSACRegressor类
# max_trials 最大迭代数100
# min_samples 随机选择样本的最低数量
# absolute_loss 作为形式参数residual_matrix的实际参数 该算法计算拟合县和采样点之间的绝对垂直距离
# residual_threshold=5.0 允许与拟合线垂直距离在5个单位以内的采样点被包括在内点集
from sklearn.linear_model import RANSACRegressor
ransac = RANSACRegressor(LinearRegression(),
                         max_trials=100, 
                         min_samples=50,
                         loss='absolute_loss',
                         residual_threshold=5.0,
                         random_state=0)
ransac.fit(X, y)
```

```python
# 默认情况下，scikit-learn 用MAD（y的平均绝对偏差） 内点阈值因问题而异
```

```python
# 根据RANSAC算法拟合的线性回归模型获得内点和离群点，绘制成图
```

```python
inlier_mask = ransac.inlier_mask_
outlier_mask = np.logical_not(inlier_mask) #逻辑 不是内点的点
```

```python
line_X = np.arange(3, 10, 1)
line_y_ransac = ransac.predict(line_X[:, np.newaxis])
plt.scatter(X[inlier_mask], y[inlier_mask],
            c='steelblue', edgecolors='white',
            marker='o', label='Inliers')
plt.scatter(X[outlier_mask], y[outlier_mask],
            c='limegreen', edgecolors='white',
            marker='s', label='Outliers')
plt.plot(line_X, line_y_ransac, color='black', lw=2)
plt.xlabel('Average number of rooms [RM]')
plt.ylabel('Price in $1000s [MEDV]')
plt.legend(loc='upper left')
plt.show()
```

## 10.7 将线性回归模型转换为曲线——多项式回归

### 10.7.1 用scikit-learn 增加多项式的项
       

```python
# scikit-learn的PolynomialFeatures转换器类，为解释变量增加二次项（d=2）
```

```python
# 1. 增加二次多项式
from sklearn.preprocessing import PolynomialFeatures
X = np.array([258.0, 270.0, 294.0, 320.0, 342.0,
              368.0, 396.0, 446.0, 480.0, 586.0])[:, np.newaxis]
y = np.array([236.4, 234.4, 252.8, 298.6, 314.2,
              342.2, 360.8, 368.0, 391.2, 390.8])
lr = LinearRegression()
pr = LinearRegression()
quadratic = PolynomialFeatures(degree=2)
X_quad = quadratic.fit_transform(X)
```

```python
# 2. 为了比较，拟合一个简单的线性回归模型
lr.fit(X, y)
X_fit = np.arange(250, 600, 10)[:, np.newaxis]
y_lin_fit = lr.predict(X_fit)
```

```python
# 3. 用转换后的特征以多项式回归拟合多元回归模型
pr.fit(X_quad, y)
y_quad_fit = pr.predict(quadratic.fit_transform(X_fit))
```

```python
# 4. 绘制结果
plt.scatter(X, y, label='training points')
plt.plot(X_fit, y_lin_fit,
         label='linear fit', linestyle='--')
plt.plot(X_fit, y_quad_fit, 
         label='quadratic fit')
plt.legend(loc='upper left')
plt.show()
```

### 10.7.2 为住房数据集中的非线性关系建模

```python
X = df[['LSTAT']].values
y = df['MEDV'].values
regr = LinearRegression()
```

```python
# 创建二次多项式
quadratic = PolynomialFeatures(degree=2)
cubic = PolynomialFeatures(degree=3)
X_quad = quadratic.fit_transform(X)
X_cubic = cubic.fit_transform(X)
```

```python
# 模型学习
X_fit = np.arange(X.min(), X.max(), 1)[:, np.newaxis]

regr = regr.fit(X, y)
y_lin_fit = regr.predict(X_fit)
linear_r2 = r2_score(y, regr.predict(X))

regr = regr.fit(X_quad, y)
y_quad_fit = regr.predict(quadratic.fit_transform(X_fit))
quadratic_r2 = r2_score(y, regr.predict(X_quad))

regr = regr.fit(X_cubic, y)
y_cubic_fit = regr.predict(cubic.fit_transform(X_fit))
cubic_r2 = r2_score(y, regr.predict(X_cubic))
```

```python
# 绘制结果
plt.scatter(X, y, label='training points', color='lightgray')

plt.plot(X_fit, y_lin_fit, 
         label='linear (d=1), $R^2=%.2f$' % linear_r2,
         color='blue', 
         lw=2,
         linestyle=':')

plt.plot(X_fit, y_quad_fit,
         label='quadratic (d=2), $R^2=%.2f$' % quadratic_r2,
         color='red',
         lw=2,
         linestyle='-')
plt.plot(X_fit, y_cubic_fit,
         label='cubic (d=3), $R^2=%.2f$' % cubic_r2,
         color='green',
         lw=2,
         linestyle='--')
plt.xlabel('% lower status of the population [LSTAT]')
plt.ylabel('Price in $1000s [MEDV]')
plt.legend(loc='upper right')
plt.show()
```
