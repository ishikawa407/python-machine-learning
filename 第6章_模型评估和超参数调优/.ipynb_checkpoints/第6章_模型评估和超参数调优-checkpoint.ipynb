```python
import pandas as pd
```

```python
import numpy as np
```

```python
df = pd.read_csv('wdbc.data', header=None)
```

```python
df
```

```python
X = df.iloc[:, 2:].values # 加上Values
```

```python
X
```

```python
y = df.iloc[:, 1].values
```

```python
y
```

```python
# 利用LabelEncoder对象将分类对象标签从原来的字符串型（‘M’， ‘B‘）转换为整数型
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
y = le.fit_transform(y)
```

```python
le.classes_
```

```python
y # 这种方法母数据df中的y标签不会变
```

```python
# 按照8：2的比例分成独立的训练集和测试集
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = \
        train_test_split(X, y, test_size=0.2, stratify=y, random_state=1)
```

```python
# 将StandardScaler, PCA, LogisticRegression 在管道中链接起来
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.pipeline import make_pipeline
```

```python
pipe_lr = make_pipeline(StandardScaler(),
                      PCA(n_components=2),
                      LogisticRegression(random_state=1))
```

```python
pipe_lr.fit(X_train, y_train)
```

```python
y_pred = pipe_lr.predict(X_test)
```

```python
print('Test Accuracy: % .3f ' % (pipe_lr.score(X_test, y_test)))
```

```python
"""
k折交叉验证评估模型的性能
    1 抵抗交叉验证
        原始数据集分为 训练数据集、交叉验证数据集、测试集
        缺点：性能评估结果可能会对如何把训练集分裂成训练和验证子集
             非常敏感，评估结果会随不同数据样本而变化
    2 k折交叉验证
        将训练集随即分裂成k个无更换子集，其中k-1个子集用于模型训练，一个子集用
        于性能评估。重复k次，得到k个模型和k次性能估计
        
        计算基于不同且独立的数据子集模型的平均性能
        
        一般k最优标准值为10
            如果训练集的规模相对较小，增加分区可能有益。
            数据集较大，可以选择较小的k值如5 可iyi完成对模型平均性能的准确评估， 降低计算成本
        
        留存交叉验证（LOOCV）
            把k值设置为训练子集的样本数（k=n）规模非常小的数据集适用
        
"""
```

```python
import numpy as np
```

```python
from sklearn.model_selection import StratifiedKFold
```

```python
kfold = StratifiedKFold(n_splits=10,
                       random_state=1).split(X_train, y_train)
```

```python
scores = []
```

```python
for k, (train, test) in enumerate(kfold):
    pipe_lr.fit(X_train[train], y_train[train])
    score = pipe_lr.score(X_train[test], y_train[test])
    scores.append(score)
    print('Fold: %2d, class dist.: %s, Acc: %.3f' % (k+1,
                                                    np.bincount(y_train[train]), score))
```

```python
print('\n CV accuracy: %.3f +/- %.3f' % (np.mean(scores), np.std(scores)))
```

```python
# scikit-learn 也实现了k折交叉验证得分器
# 可以更简洁的评估模型
```

```python
from sklearn.model_selection import cross_val_score
```

```python
scores = cross_val_score(estimator=pipe_lr,
                         X=X_train,
                         y=y_train,
                         cv=10,
                         n_jobs=1) # n-jobs 可以分配计算机不同CPU计算 -1为所有CPU
print('CV accuracy scores: %s' % scores)
print('\n CV accuracy: %.3f +/- %.3f' % (np.mean(scores), np.std(scores)))
```

```python
# 6.3.1 用学习曲线诊断偏差和方差问题
```

```python
"""
高偏差
    表现：    模型的训练和交叉验证准确度均低，说明模型对训练数据欠拟合
    解决方法: 增加模型的参数个数，模型复杂度
                收集构建额外的特征，降低正则化成都
高方差
    表现：    模型在训练和交叉验证的准确度上有较大的差别
    解决方法： 收集更多的训练数据，降低模型复杂度，增加正则化参数
    
"""
```

```python
# 利用scikit-learn学习曲线进行模型评估
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import learning_curve
```

```python
pipe_lr = make_pipeline(StandardScaler(),
                       LogisticRegression(penalty='l2',
                                         random_state=1))
```

```python
train_sizes, train_scores, test_scores = \
                learning_curve(estimator=pipe_lr,
                               X=X_train,
                               y=y_train,
                               train_sizes=np.linspace(
                                   0.1, 1.0, 10), # np.linspace() 0.1-10 分10个
                               cv=10,
                               n_jobs=-1)
```

```python
train_mean = np.mean(train_scores, axis=1)
train_std = np.std(train_scores, axis=1)
test_mean = np.mean(test_scores, axis=1)
test_std = np.std(test_scores, axis=1)
```

```python
plt.plot(train_sizes, train_mean,
        color='blue', marker='o',
        markersize=5, label='training accuracy')
```

```python
plt.plot(train_sizes, train_mean,
        color='blue', marker='o',
        markersize=5, label='training accuracy')
plt.fill_between(train_sizes,
                train_mean + train_std,
                train_mean - train_std,
                alpha=0.15, color='blue')
```

```python
plt.plot(train_sizes, train_mean,
        color='blue', marker='o',
        markersize=5, label='training accuracy')
plt.fill_between(train_sizes,
                train_mean + train_std,
                train_mean - train_std,
                alpha=0.15, color='blue')
plt.plot(train_sizes, test_mean,
        color='green', linestyle='--',
        marker='s', markersize=5,
        label='validataion accuracy')
```

```python
plt.plot(train_sizes, train_mean,
        color='blue', marker='o',
        markersize=5, label='training accuracy')
plt.fill_between(train_sizes,
                train_mean + train_std,
                train_mean - train_std,
                alpha=0.15, color='blue')
plt.plot(train_sizes, test_mean,
        color='green', linestyle='--',
        marker='s', markersize=5,
        label='validataion accuracy')
plt.fill_between(train_sizes,
                test_mean + test_std,
                test_mean - test_std,
                alpha=0.15, color='green')
plt.grid()
plt.xlabel('Number of training samples')
plt.ylabel('Accuracy')
plt.legend(loc='lower right')
plt.ylim([0.8, 1.03])
plt.show()
```

```python
param_range = [0.001, 0.01, 0.1, 1.0, 10.0, 100.0]
```

```python
train_std = np.std(train_scores, axis=1)
```

```python
test_std = np.std(test_scores, axis=1)
```

```python
plt.plot(param_range, train_mean,
        color='blue', marker='o',
        markersize=5, label='training accuracy')
plt.fill_between(param_range, train_mean + train_std,
                train_mean - train_std, alpha=0.15,
                color='blue')
plt.plot(param_range, test_mean,
        color='green', linestyle='--',
        marker='s', markersize=5,
        label='validation accuracy')
plt.fill_between(param_range,
                test_mean + test_std,
                test_mean - test_std,
                alpha=0.15, color='green')
plt.grid()
plt.xscale('log')
plt.legend(loc='lower right')
plt.xlabel('Parameter C')
plt.ylabel('Accuracy')
plt.ylim([0.8, 1.03])
plt.show()
```

```python
"""
6.4 通过网格搜索为机器学习模型调优

    机器学习两类参数：
        1. 从训练数据中学习到的参数，例如逻辑回归的权重；
        2. 单独优化的算法参数，为模型的调优参数，也称超参数，例如正则化参数
    
    网格搜索：为常见超参数调优技术，寻找最优超参数组合进一步提高模型性能
"""
```

```python
# 6.4.1 通过网格搜索为超参数调优
"""
    网格搜索法逻辑：属于暴力穷举搜索类型，预先定义好不同的超参数值，\
                  计算机针对每种组合分别评估模型的性能，获得最佳 \
                  组合参数值
"""
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.svm import SVC
```

```python
pipe_svc = make_pipeline(StandardScaler(),
                        SVC(random_state=1))
```

```python
param_range = [0.0001, 0.001, 0.01, 0.1,
              1.0, 10.0, 100.0, 1000.0]
```

```python
param_grid = [{'svc__C': param_range,
               'svc__kernel': ['linear']},
              {'svc__C': param_range,
               'svc__gamma': param_range,
               'svc__kernel': ['rbf']}]
```

```python
gs = GridSearchCV(estimator=pipe_svc,
                  param_grid=param_grid,　　# 设置为字典的列表
                  scoring='accuracy',
                  cv=10,
                  n_jobs=-1)
```

```python
gs = gs.fit(X_train, y_train)
```

```python
print(gs.best_score_)
```

```python
print(gs.best_params_)
```

```python
# 用独立的测试集评估所选最佳模型的性能
clf = gs.best_estimator_
clf.fit(X_train, y_train) #  这一步可以省
print('Test accuracy: %.3f' % clf.score(X_test, y_test))
```

```python
gs.best_estimator_.score(X_test, y_test)
```

```python
clf.score(X_test, y_test)
```
