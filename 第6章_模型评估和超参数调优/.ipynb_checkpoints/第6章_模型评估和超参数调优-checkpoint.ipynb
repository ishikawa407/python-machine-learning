```python
import pandas as pd
```

```python
import numpy as np
```

```python
df = pd.read_csv('wdbc.data', header=None)
```

```python
df
```

```python
X = df.iloc[:, 2:].values # åŠ ä¸ŠValues
```

```python
X
```

```python
y = df.iloc[:, 1].values
```

```python
y
```

```python
# åˆ©ç”¨LabelEncoderå¯¹è±¡å°†åˆ†ç±»å¯¹è±¡æ ‡ç­¾ä»ŽåŽŸæ¥çš„å­—ç¬¦ä¸²åž‹ï¼ˆâ€˜Mâ€™ï¼Œ â€˜Bâ€˜ï¼‰è½¬æ¢ä¸ºæ•´æ•°åž‹
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
y = le.fit_transform(y)
```

```python
le.classes_
```

```python
y # è¿™ç§æ–¹æ³•æ¯æ•°æ®dfä¸­çš„yæ ‡ç­¾ä¸ä¼šå˜
```

```python
# æŒ‰ç…§8ï¼š2çš„æ¯”ä¾‹åˆ†æˆç‹¬ç«‹çš„è®­ç»ƒé›†å’Œæµ‹è¯•é›†
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = \
        train_test_split(X, y, test_size=0.2, stratify=y, random_state=1)
```

```python
# å°†StandardScaler, PCA, LogisticRegression åœ¨ç®¡é“ä¸­é“¾æŽ¥èµ·æ¥
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.pipeline import make_pipeline
```

```python
pipe_lr = make_pipeline(StandardScaler(),
                      PCA(n_components=2),
                      LogisticRegression(random_state=1))
```

```python
pipe_lr.fit(X_train, y_train)
```

```python
y_pred = pipe_lr.predict(X_test)
```

```python
print('Test Accuracy: % .3f ' % (pipe_lr.score(X_test, y_test)))
```

```python
"""
kæŠ˜äº¤å‰éªŒè¯è¯„ä¼°æ¨¡åž‹çš„æ€§èƒ½
    1 æŠµæŠ—äº¤å‰éªŒè¯
        åŽŸå§‹æ•°æ®é›†åˆ†ä¸º è®­ç»ƒæ•°æ®é›†ã€äº¤å‰éªŒè¯æ•°æ®é›†ã€æµ‹è¯•é›†
        ç¼ºç‚¹ï¼šæ€§èƒ½è¯„ä¼°ç»“æžœå¯èƒ½ä¼šå¯¹å¦‚ä½•æŠŠè®­ç»ƒé›†åˆ†è£‚æˆè®­ç»ƒå’ŒéªŒè¯å­é›†
             éžå¸¸æ•æ„Ÿï¼Œè¯„ä¼°ç»“æžœä¼šéšä¸åŒæ•°æ®æ ·æœ¬è€Œå˜åŒ–
    2 kæŠ˜äº¤å‰éªŒè¯
        å°†è®­ç»ƒé›†éšå³åˆ†è£‚æˆkä¸ªæ— æ›´æ¢å­é›†ï¼Œå…¶ä¸­k-1ä¸ªå­é›†ç”¨äºŽæ¨¡åž‹è®­ç»ƒï¼Œä¸€ä¸ªå­é›†ç”¨
        äºŽæ€§èƒ½è¯„ä¼°ã€‚é‡å¤kæ¬¡ï¼Œå¾—åˆ°kä¸ªæ¨¡åž‹å’Œkæ¬¡æ€§èƒ½ä¼°è®¡
        
        è®¡ç®—åŸºäºŽä¸åŒä¸”ç‹¬ç«‹çš„æ•°æ®å­é›†æ¨¡åž‹çš„å¹³å‡æ€§èƒ½
        
        ä¸€èˆ¬kæœ€ä¼˜æ ‡å‡†å€¼ä¸º10
            å¦‚æžœè®­ç»ƒé›†çš„è§„æ¨¡ç›¸å¯¹è¾ƒå°ï¼Œå¢žåŠ åˆ†åŒºå¯èƒ½æœ‰ç›Šã€‚
            æ•°æ®é›†è¾ƒå¤§ï¼Œå¯ä»¥é€‰æ‹©è¾ƒå°çš„kå€¼å¦‚5 å¯iyiå®Œæˆå¯¹æ¨¡åž‹å¹³å‡æ€§èƒ½çš„å‡†ç¡®è¯„ä¼°ï¼Œ é™ä½Žè®¡ç®—æˆæœ¬
        
        ç•™å­˜äº¤å‰éªŒè¯ï¼ˆLOOCVï¼‰
            æŠŠkå€¼è®¾ç½®ä¸ºè®­ç»ƒå­é›†çš„æ ·æœ¬æ•°ï¼ˆk=nï¼‰è§„æ¨¡éžå¸¸å°çš„æ•°æ®é›†é€‚ç”¨
        
"""
```

```python
import numpy as np
```

```python
from sklearn.model_selection import StratifiedKFold
```

```python
kfold = StratifiedKFold(n_splits=10,
                       random_state=1).split(X_train, y_train)
```

```python
scores = []
```

```python
for k, (train, test) in enumerate(kfold):
    pipe_lr.fit(X_train[train], y_train[train])
    score = pipe_lr.score(X_train[test], y_train[test])
    scores.append(score)
    print('Fold: %2d, class dist.: %s, Acc: %.3f' % (k+1,
                                                    np.bincount(y_train[train]), score))
```

```python
print('\n CV accuracy: %.3f +/- %.3f' % (np.mean(scores), np.std(scores)))
```

```python
# scikit-learn ä¹Ÿå®žçŽ°äº†kæŠ˜äº¤å‰éªŒè¯å¾—åˆ†å™¨
# å¯ä»¥æ›´ç®€æ´çš„è¯„ä¼°æ¨¡åž‹
```

```python
from sklearn.model_selection import cross_val_score
```

```python
scores = cross_val_score(estimator=pipe_lr,
                         X=X_train,
                         y=y_train,
                         cv=10,
                         n_jobs=1) # n-jobs å¯ä»¥åˆ†é…è®¡ç®—æœºä¸åŒCPUè®¡ç®— -1ä¸ºæ‰€æœ‰CPU
print('CV accuracy scores: %s' % scores)
print('\n CV accuracy: %.3f +/- %.3f' % (np.mean(scores), np.std(scores)))
```

```python
# åˆ©ç”¨scikit-learnå­¦ä¹ æ›²çº¿è¿›è¡Œæ¨¡åž‹è¯„ä¼°
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import learning_curve
```

```python
pipe_lr = make_pipeline(StandardScaler(),
                       LogisticRegression(penalty='l2',
                                         random_state=1))
```

```python
train_sizes, train_scores, test_scores = \
                learning_curve(estimator=pipe_lr,
                               X=X_train,
                               y=y_train,
                               train_sizes=np.linspace(
                                   0.1, 1.0, 10), # np.linspace() 0.1-10 åˆ†10ä¸ª
                               cv=10,
                               n_jobs=-1)
```

```python
train_mean = np.mean(train_scores, axis=1)
train_std = np.std(train_scores, axis=1)
test_mean = np.mean(test_scores, axis=1)
test_std = np.std(test_scores, axis=1)
```

```python
plt.plot(train_sizes, train_mean,
        color='blue', marker='o',
        markersize=5, label='training accuracy')
```

```python
plt.plot(train_sizes, train_mean,
        color='blue', marker='o',
        markersize=5, label='training accuracy')
plt.fill_between(train_sizes,
                train_mean + train_std,
                train_mean - train_std,
                alpha=0.15, color='blue')
```

```python
plt.plot(train_sizes, train_mean,
        color='blue', marker='o',
        markersize=5, label='training accuracy')
plt.fill_between(train_sizes,
                train_mean + train_std,
                train_mean - train_std,
                alpha=0.15, color='blue')
plt.plot(train_sizes, test_mean,
        color='green', linestyle='--',
        marker='s', markersize=5,
        label='validataion accuracy')
```

```python
plt.plot(train_sizes, train_mean,
        color='blue', marker='o',
        markersize=5, label='training accuracy')
plt.fill_between(train_sizes,
                train_mean + train_std,
                train_mean - train_std,
                alpha=0.15, color='blue')
plt.plot(train_sizes, test_mean,
        color='green', linestyle='--',
        marker='s', markersize=5,
        label='validataion accuracy')
plt.fill_between(train_sizes,
                test_mean + test_std,
                test_mean - test_std,
                alpha=0.15, color='green')
plt.grid()
plt.xlabel('Number of training samples')
plt.ylabel('Accuracy')
plt.legend(loc='lower right')
plt.ylim([0.8, 1.03])
plt.show()
```

```python
from sklearn.model_selection import validation_curve
```

```python
param_range = [0.001, 0.01, 0.1, 1.0, 10.0, 100.0]
```

```python
train_scores, test_scores = validation_curve(
                estimator=pipe_lr,
                X=X_train,
                y=y_train,
                param_name='logisticregression__C',# ä¸èƒ½è¾“é”™
                param_range=param_range,
                cv=10,
                n_jobs=-1)
```

```python
train_mean = np.mean(train_scores, axis=1)
```

```python
train_std = np.std(train_scores, axis=1)
```

```python
test_mean = np.mean(test_scores, axis=1)
```

```python
test_std = np.std(test_scores, axis=1)
```

```python
plt.plot(param_range, train_mean,
        color='blue', marker='o',
        markersize=5, label='training accuracy')
plt.fill_between(param_range, train_mean + train_std,
                train_mean - train_std, alpha=0.15,
                color='blue')
plt.plot(param_range, test_mean,
        color='green', linestyle='--',
        marker='s', markersize=5,
        label='validation accuracy')
plt.fill_between(param_range,
                test_mean + test_std,
                test_mean - test_std,
                alpha=0.15, color='green')
plt.grid()
plt.xscale('log')
plt.legend(loc='lower right')
plt.xlabel('Parameter C')
plt.ylabel('Accuracy')
plt.ylim([0.8, 1.03])
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.svm import SVC
```

```python
pipe_svc = make_pipeline(StandardScaler(),
                        SVC(random_state=1))
```

```python
param_range = [0.0001, 0.001, 0.01, 0.1,
              1.0, 10.0, 100.0, 1000.0]
```

```python
param_grid = [{'svc__C': param_range,
               'svc__kernel': ['linear']},
              {'svc__C': param_range,
               'svc__gamma': param_range,
               'svc__kernel': ['rbf']}]
```

```python
gs = GridSearchCV(estimator=pipe_svc,
                  param_grid=param_grid,ã€€ã€€# è®¾ç½®ä¸ºå­—å…¸çš„åˆ—è¡¨
                  scoring='accuracy',
                  cv=10,
                  n_jobs=-1)
```

```python
gs = gs.fit(X_train, y_train)
```

```python
print(gs.best_score_)
```

```python
print(gs.best_params_)
```

```python
# ç”¨ç‹¬ç«‹çš„æµ‹è¯•é›†è¯„ä¼°æ‰€é€‰æœ€ä½³æ¨¡åž‹çš„æ€§èƒ½
clf = gs.best_estimator_
clf.fit(X_train, y_train) #  è¿™ä¸€æ­¥å¯ä»¥çœ
print('Test accuracy: %.3f' % clf.score(X_test, y_test))
```
